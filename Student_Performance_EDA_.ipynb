```python
students_grades_df = students_grades_df.drop(['G1','G2','G3'], axis = 1)
students_grades_df
```

```python
def detect_outliers(columns):
  outlier_indices = []
  for col in columns:
    Q1 = students_grades_df[col].quantile(0.25)
    Q3 = students_grades_df[col].quantile(0.75)
    IQR = Q3-Q1

    mask = (students_grades_df[col]>= (Q1-1.5)*IQR) & (students_grades_df[col] <= (Q1+1.5)*IQR)

    mask = mask.to_numpy()
    false_indices = np.argwhere(~mask)
    outlier_indices.append(false_indices)
    return np.unique(np.concatenate(outlier_indices).ravel())
```

```python
num_col= ["age","absence"]
detect_outliers(num_col)
```
